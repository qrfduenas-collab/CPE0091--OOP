  IF STATEMENT

```python
a=12
b=100
if b>a:
  print("b is greater than a")
```

ELIF STATEMENTS

```python
a=12
b=100
if b>a:
  print("b is greater than a")

elif a==b:
    print("a and b are equal")
```

  ELSE STATEMENTS

```python
a=200
b=33
if b>a:
  print("b is greater than a")
elif b==a:
  print ("b is equal to a")
else:
  print ("a is greater than b")
```

```python
a=22
b=33
c=500
if a > b and c > a:
  print("Both conditions are true")
```

NESTED IF

```python
print("Enter a number")
x = int(input())  # user input

if x > 10:
    print("Above ten,")
    if x > 20:
        print("and also above 20!")
        if x > 30:
            print("and also above 30")
            if x > 50:
                print("and also above 50")
            else:
                print("below 50")
        else:
            print("below 30")
    else:
        print("below 20.")
else:
    print("below 10")
```
